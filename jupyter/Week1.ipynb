```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

elite = pd.read_csv('../inputs/boardgame-elite-users.csv')
frequent = pd.read_csv('../inputs/boardgame-frequent-users.csv')
titles = pd.read_csv('../inputs/boardgame-titles.csv')
elite = elite.rename(columns = {'Compiled from boardgamegeek.com by Matt Borthwick':'UserID'})
frequent = frequent.rename(columns = {'Compiled from boardgamegeek.com by Matt Borthwick':'UserID'})
titles = titles.rename(columns={'boardgamegeek.com game ID':'gameID'})

elite.rating.hist()
```

```python
agg = frequent.groupby('gameID',as_index=False).agg(np.mean)
agg = agg.set_index('gameID').join(titles.set_index('gameID'))
agg.sort_values('rating',ascending=False)
```

```python
#def pre_process(X):
DAMPENING_TERM = 25

pd.set_option('display.max_rows',1000)
mean = elite.rating.mean()
R = elite.pivot_table(index='UserID',columns='gameID',values='rating')

def user_base(row, mean, damping=False):
    bu = row.mean() - mean
    if damping:
        num_items_user_reviewed = row[row.notnull()].size
        damp_factor = num_items_user_reviewed/(num_items_user_reviewed+DAMPENING_TERM)
        bu*=damp_factor
    return bu


def item_base(column, mean, bu, damping=False):
    
    users_that_reviewed_this_item = column[column.notnull()]
    bu_for_users_that_reviewed_i = bu[users_that_reviewed_this_item.index].mean()
    bi = users_that_reviewed_this_item.mean()-bu_for_users_that_reviewed_i-mean
    if damping:
        num_users_reviewed_item = column[column.notnull()].size
        damp_factor = num_users_reviewed_item/(num_users_reviewed_item+DAMPENING_TERM)
        bi*=damp_factor
    return bi
    

bu = R.apply(lambda x: user_base(x,mean), axis=1)
bi = R.apply(lambda x: item_base(x,mean,bu))

def predict(user,item,bu,bi,mean):
    return mean + bu[user] + bi[item]

pred = predict(272,11,bu,bi,mean)
print(R.loc[272,11],pred)

```
